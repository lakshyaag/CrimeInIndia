```r
pacman::p_load('dplyr', 'tidyverse', 'ggplot2',  'ggalt', 'forcats', 'R.utils', 'png', 'grid', 'ggpubr', 'scales', 'bbplot', 'ggthemes', 'plotly')

crime <- read.csv('crime_2016_tidy.csv')

# List of crimes
unique(crime$crime)

plot_crime <- function(crime_type, color){
    plot <- ggplot(crime %>% filter(crime == crime_type), aes(x=reorder(state, incidents), y=incidents)) + 
        geom_col(fill=color) + 
        geom_hline(yintercept = 0, size = 1, colour="#333333") +
        guides(fill=F) +
        bbc_style() +
        labs(title=paste('Incidents of', crime_type), subtitle = 'Crime Statistics, 2016', y='Number of incidents') +
        coord_flip() +
        scale_y_continuous(labels = scales::comma) +
        theme(axis.text.x = element_text(margin=margin(t = 10, b = 10)))
    
    finalise_plot(plot, source = 'Source: National Crime Record Bureau', save_filepath = paste('outputs/', crime_type, '.jpg', sep = ''), width_pixels = 1000, height_pixels = 1000)
}

plot_crime('Rape- Rape-Sec 376 IPC', '#d31234')
```

```r
plot_state <- function(state_name, color){
    state_crime_data <- crime %>% filter(state == 'Uttar Pradesh' & !crime %in% c('Total Cognizable IPC Crime', 'Other IPC Crimes')) %>% top_n(10, wt = incidents) %>%
        mutate(crime = str_extract(crime, '(Sec.*IPC)'))
    plot <- ggplot(state_crime_data, aes(x=reorder(crime, incidents), y=incidents)) + 
        geom_col(fill=color) + 
        geom_hline(yintercept = 0, size = 1, colour="#333333") +
        guides(fill=F) +
        bbc_style() +
        labs(title=paste('Incidents in', state_name), subtitle = 'Crime Statistics, 2016', y='Number of incidents') +
        scale_y_continuous(labels = scales::comma) +
        coord_flip()
    
    finalise_plot(plot, source = 'Source: National Crime Record Bureau', save_filepath = paste('outputs/', state_name, '.jpg', sep = ''), width_pixels = 1000, height_pixels = 1000)
}

plot_state('Uttar Pradesh', '#d32341')
```

```r
overview_plot <- ggplot(crime %>% group_by(state) %>% filter(crime == 'Total Cognizable IPC Crime'), aes(x=reorder(state, -incidents), y=incidents)) + 
        geom_col(fill='#F63231') + 
        geom_hline(yintercept = 0, size = 1, colour="#333333") +
        guides(fill=F) +
        bbc_style() +
        labs(title='Overview of crimes', subtitle = 'Crime Statistics, 2016', y='Number of incidents') +
        scale_y_continuous(labels = scales::comma, breaks = c(0, 50000, 100000, 150000, 200000, 250000)) +
        scale_x_discrete(labels=abbreviate) +
        theme(axis.text.x = element_text(angle = 90))

finalise_plot(overview_plot, source = 'Source: National Crime Record Bureau', save_filepath = 'outputs/overview_plot.jpg', width_pixels = 1000, height_pixels = 1000)
```
